Plot the population size of each species for each year on a single graph. Add legend and axis labels to the plot.



```python
import matplotlib.pyplot as plt
import numpy as np
data = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt')
fig = plt.figure(figsize=(16, 9))
fig.patch.set_facecolor('xkcd:white')
plt.xticks(range(1900, 1921))
D = data[:, 0]
H = data[:, 1]
L = data[:, 2]
C = data[:, 3]
plt.plot(D, H, label='hare', c = 'green', linestyle='--')
plt.plot(D, L, label='lynx', c ='red', linewidth=1)
plt.plot(D, C, label='carrot', c ='orange', linewidth=2)
plt.xlabel('YEARS', fontsize=15)
plt.ylabel('POPULATION')
plt.xticks(fontsize=7)
plt.legend(loc = (1.02, 0.82))
plt.show()
```

Задача: написать функцию, реализующую алгоритм kNN.

Входом в функцию является

train_X - матрица признаков обучающего датасета
train_Y - массив классов обучающего датасета
test_X - матрица признаков тестового датасета
k - число соседей
Выход:

массив той же длины, что и test_X, содержащий классы для тестового датасета
Фунция должна содержать не более 1 цикла for/while! Стремитесь обойтись как можно меньшим количеством строк кода.

Примените kNN на выборкe x_test из датасета MNIST для разных K. Посчитать точность классификации (accuracy - доля правильно классифицированных объектов) и выбрать оптимальное К.



```python
import numpy as np


def KNN(train_X, train_Y, test_X, k):
    res = []
    for testi in range(len(test_X)): #элементы из test_X
        dist = np.sqrt(np.sum((train_X - test_X[testi])**2, axis=1))##расстояние между элемнтом test_X и элемантами train_X
        neark = np.argsort(dist)[:k] # Находим индексы k ближайших соседей
        nearkmark = train_Y[neark]# маркеры классов для k ближайших соседей
        testmark, numkmark = np.unique(nearkmark, return_counts=True)##маркеры классов и их количество среди k ближайших соседей
        res.append(testmark[np.argmax(numkmark)])## предсказанный класс для текущего элемента test_X
    return np.array(res) # список предсказанных классов для test_X
```

```python
from sklearn. datasets import fetch_openml
#Загрузка может занять около минуты
images, labels = fetch_openml('mnist_784', return_X_y=True, as_frame=False)
```

```python
trainsize = 1000
testsize = 100
trainimages = images [:trainsize]
testimages = images[-testsize:]
trainlabels = labels [:trainsize]
testlabels = labels [-testsize:]
```

```python
presk = 1
presaccuracy = 0
import sklearn
for k in range(1, 1001):
    model = KNN(trainimages, trainlabels, testimages, k)
    from sklearn import metrics
    accuracy = sklearn.metrics.accuracy_score(testlabels, model)
    if accuracy > presaccuracy:
        presaccuracy = accuracy
        presk = k
```

```python
print("Оптимальное значение K:", presk)
print("Точность классификации:", presaccuracy)
```
