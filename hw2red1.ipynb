Создайте массив 10х10, заполненный в центре нулями, а по границе единицами.
(было без цикла, поэтому без исправлений)


```python
import numpy as np
x = np.ones((10, 10))
x[1:-1, 1:-1] = 0
print(x)
```

Создайте массив 8х8, заполненный 0 и 1 в шахматном порядке.
(было без цикла, поэтому без исправлений)

```python
x = np.ones([8,8])
x[1::2, ::2] = 0
x[::2, 1::2] = 0
print(x)
```

Создайте случайный массив 5х5 и нормируйте его так, чтобы среднее по каждой строке было в точности ноль.(исправлено)

```python
x = np.random.random((5,5))
nx = x - np.mean(x, axis =1, keepdims = True)
print(nx)
```

Создайте двумерный массив 10x3, заполненный случайными числами от 0 до 1. В каждой строке выберите значение, наиболее близкое к 0.5.(исправлено)

```python
x = np.random.sample((10,3))
print(x)
print ([x[i ,np.argmin(abs(x - 0.5), axis = 1)[i]] for i in range(x.shape[0])])
```

Создайте двумерный массив 6x6, заполненный случайными значениями. Сумму по каждой строке поделите на максимальный элемент из столбца с тем же индексом.
(исправлено)

На примере массива x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) найдите максимальный элемент в массиве среди элементов, перед которыми стоит ноль.(как было)

```python
x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])
print(max((x[i] for i in range(1, len(x)) if x[i - 1] == 0)))
```

Пусть заданы два массива x = np.ones(10) и i = np.array([0, 1, 2, 3, 5, 5, 5, 8]). Прибавьте единицу к тем элементам массива x, индексы которых указаны в массиве i. В случае, если некоторый индекс встретился в массиве i несколько раз, прибавьте к соответствующему элементу массива x число вхождений данного индекса в массив i.(исправлено)

```python
x = np.ones(10)
i = np.array([0, 1, 2, 3, 5, 5, 5, 8])
x += np.bincount(i, minlength = len(x))
print(x)
```
