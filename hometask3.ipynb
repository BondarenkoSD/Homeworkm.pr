Plot the population size of each species for each year on a single graph. Add legend and axis labels to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt')
year, hares, lynxes, carrots = data.T
plt.axes([0.5, 0.5, 2., 1.])
plt.plot(year, hares, year, lynxes, year, carrots)
##plt.xticks(np.arange(min(year), max(year)+1, 1.0))  #      X
plt.xlabel('YEAR', fontsize = 15)
plt.ylabel('POPULATION', fontsize = 15)
plt.legend(('Hare', 'Lynx', 'Carrot'), loc = (1.02, 0.82))
plt.show()
```

Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for each year in the period.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sb
from matplotlib import rcParams
import pandas as pd
import numpy as np
from math import *
data = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt').transpose()
plt.figure(figsize = (16, 8))
plt.rcParams['font.size'] = '10'
plt.xlabel('YEAR', fontsize = 15)
plt.ylabel('POPULATION PERCENT', fontsize = 15)
stack_dict = {data[0][i] : [floor((data[1][i]/(data[1][i] + data[2][i])) * 100)/100, (ceil(data[2][i]/(data[1][i] + data[2][i])*100))/100] for i in range(len(data[0]))}
for i in range(len(data[0])):
    plt.bar(data[0], data[1], width = 0.9)
    plt.text(data[0][i]-0.385, data[1][i]/2 - 1000, f'{int(stack_dict[data[0][i]][0]*100)}%')
    plt.bar(data[0], data[2], bottom = data[1], width = 0.9)
    plt.text(data[0][i] - 0.385, data[1][i] + data[2][i]/2 - 1000, f'{int(stack_dict[data[0][i]][1]*100)}%')
plt.legend(['Hares', 'Lynxes'])
plt.show()
```
